# Star Patterns

## 1. Triangular Star patterns :

```python
'''
1. *
   * * 
   * * *
   * * * *
   * * * * * 
'''
num = int(input("Enter the number: "))
for i in range(num):
    for j in range(num):
        if i>=j:
            print("*",end=" ")
        else:
            print(" ",end=" ")
    print()
```

```python
'''
2.        * 
        * * 
      * * * 
    * * * * 
  * * * * * 
'''

num = int(input("Enter the number: "))
for i in range(num+1):
    for j in range(num+1):
        if j>=(num+1)-i:
            print("*",end = " ")
        else:
            print(" ",end = " ")
    print()
```

```python
'''
3. * * * * *  
   * * * *    
   * * *      
   * *        
   *  
'''
num = int(input("Enter the number: "))
for i in range(1,num+1):
    for j in range(1,num+1):
        if j<=(num+1)-i:
            print("*",end=" ")
        else:
            print(" ",end=" ")
    print(" ")
```

```python
'''
4. * * * * *  
     * * * *  
       * * *  
         * *  
           *  
'''

num = int(input("Enter the number: "))
for i in range(1,num+1):
    for j in range(1,num+1):
        if j>=i:
            print("*",end=" ")
        else:
            print(" ",end=" ")
    print(" ")
```

```python
'''
5.
        *
      * * *
    * * * * *
  * * * * * * *
* * * * * * * * *
'''
row = int(input("Enter the number of row: "))
coloumn = int(input("Enter the number of coloumn: "))
for i in range(1,row+1):
    for j in range(1,coloumn+1):
        if j>=(row+1)-i or j<=3+i:
            print("*",end=" ")
        else:
            print(" ",end=" ")
    print(" ")
```
