### Data Visualization With Seaborn

##### Seaborn is a Python visualization library based on Matplotlib that provides a high-level interface for drawing attractive and informative statistical graphics. Seaborn helps in creating complex visualizations with just a few lines of code.

## Basic ploting with seaborn :

```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips
```

## Creating a Scatter Plot Using Seaborn :

```python
import matplotlib.pyplot as plt
```

```python
sns.scatterplot(x = 'total_bill', y = 'tip',data = tips)
```

## Creating a Line Plot Using Seaborn : 

```python
sns.lineplot(x = 'size', y='total_bill',data = tips)
plt.title("Line Plot of Total Bill Vs Size")
plt.show()
```

## Creating a Categorial Plot using Seaborn :

## Bar Plot :

```python
sns.barplot(x = 'day', y = 'total_bill',data = tips)
plt.title('Bar Plot of Total Bill Vs Day')
plt.show()
```

## Box Plot :

```python
sns.boxplot(x = "day",y = 'total_bill',data = tips)
plt.title('Box Plot Of Total Bill Vs Day')
plt.show()
```

## Violin Plot :

```python
sns.violinplot(x = 'day',y = 'total_bill',data = tips)
plt.title("Violin Plot Of Total Bill Vs Day")
plt.show()
```

## Histogram Plot :

##### In Seaborn's histogram plotting function, the kde parameter stands for Kernel Density Estimate. It is a statistical technique used to smooth data and estimate its probability density function (PDF).

##### kde=True: A KDE curve is drawn on top of the histogram. This curve gives a smoothed representation of the data distribution.

##### kde=False: No KDE curve is added. Only the histogram is displayed.

```python
sns.histplot(tips['total_bill'],bins=10,kde=True)
plt.title('Histogram Plot Of Total Bill Vs Day')
plt.show()
```

```python
sns.histplot(tips['total_bill'],bins=10,kde=False)
plt.title('Histogram Plot Of Total Bill Vs Day')
plt.show()
```

## KDE Plot :

```python
sns.kdeplot(tips['total_bill'],fill = True)
plt.title('KDE Plot')
plt.show()
```

## Pair Plot :

```python
sns.pairplot(tips)
```

```python
corr = tips[['total_bill','tip','size']].corr()
corr
```

```python
sns.heatmap(corr,annot = True,cmap= 'coolwarm')
```
